```python
import numpy as onp
from scipy.stats import arcsine
from jax import numpy as np
from jax import grad, jit, vmap
from jax.nn import sigmoid, relu, leaky_relu, elu
from matplotlib import pyplot as plt
```

```python
path = ''
```

```python
plt.rcParams.update(plt.rcParamsDefault)
plt.rc('font', family='serif')
plt.rcParams.update({'font.size': 16,
                        'lines.linewidth': 2,
                        'axes.labelsize': 10,
                        'axes.titlesize': 16,
                        'xtick.labelsize': 10,
                        'ytick.labelsize': 10,
                        'legend.fontsize': 10,
                        'axes.linewidth': 2})
```

```python
sigmas = np.linspace(0, 1, 100)

Xs = onp.random.randn(4096, *sigmas.shape)
```

```python
(Xs * sigmas).shape
```

```python
plt.figure(figsize = (4,3))

plt.plot(sigmas, np.sqrt((np.tanh(Xs*sigmas)**2).mean(0)), label = 'Tanh')
plt.plot(sigmas, np.sqrt((sigmoid(Xs*sigmas)**2).mean(0)), label = 'Sigmoid')
plt.plot(sigmas, np.sqrt((relu(Xs*sigmas)**2).mean(0)), label = 'ReLu')
plt.plot(sigmas, np.sqrt(2 * (relu(Xs*sigmas)**2).mean(0)), label = 'ReLu, He')

plt.plot(sigmas, np.sqrt((np.sin(np.pi / 2 * np.sqrt(2) * Xs * sigmas) ** 2).mean(0)), label = 'SIREN')
plt.axvline(np.sqrt(0.5), linestyle = '--', color = 'k', linewidth = 1, label = 'SIREN Var')

plt.plot(sigmas, sigmas, 'k--', alpha = 0.5, label = 'Identity')
plt.xlabel(r'$s_i$')
plt.ylabel(r'$s_{i+1}$')
plt.ylim([0., 2])
plt.legend(ncol=2)
plt.tight_layout()

plt.savefig(path + 'dynamical_system.pdf', dpi = 200)
# plt.xscale('log')
# plt.yscale('log')
plt.show()
plt.close()
```

```python
def std_mapping(activation):
    @jit
    def map_fn(sigma, X, offset = 1.):
        '''
        X is unit normal
        '''
        
        return np.sqrt((activation(offset * X * sigma)**2).mean(0))
    return map_fn

X = np.array(onp.random.randn(4096, *sigmas.shape))
plt.figure(figsize = (5,3))

for activation in [np.tanh, sigmoid, relu]:
    mapping = std_mapping(activation)
    map_grad = grad(mapping)
    map_dev = vmap(map_grad, in_axes = (0, 1))(sigmas, X)
    plt.plot(sigmas, map_dev, label = f'{activation.__name__}')

mapping = std_mapping(relu)
map_grad = lambda sigma, X : grad(mapping)(sigma, X, np.sqrt(2))
map_dev = vmap(map_grad, in_axes = (0, 1))(sigmas, X)
plt.plot(sigmas, map_dev, label = 'relu, He')

mapping = std_mapping(lambda x: np.sin(np.pi / 2 * x))
map_grad = lambda sigma, X : grad(mapping)(sigma, X, np.sqrt(2))
map_dev = vmap(map_grad, in_axes = (0, 1))(sigmas, X)
plt.plot(sigmas, map_dev, label = 'SIREN')

plt.axvline(np.sqrt(0.5), linestyle = '--', color = 'k', linewidth = 1, label = 'SIREN Var')

plt.axhline(1, linestyle = '-.', color = 'k', alpha = 0.5)
plt.legend(bbox_to_anchor = (1., 1.))
plt.axhline(np.sqrt(0.5), linestyle = '-.', color = 'k', alpha = 0.5, linewidth = 2)
plt.yticks(list(plt.yticks()[0]) + [np.sqrt(0.5)])



plt.xlabel(r'$s_i$')
plt.ylabel(r'$\frac{d}{d s_{i}}s_{i+1}$')
plt.tight_layout()
plt.savefig(path + 'rhs_derivative.pdf', dpi = 200)
plt.show()
plt.close()
```

```python
plt.figure(figsize = (4,3))
for c in [1., 2., 3.]:
    plt.plot(sigmas, np.sqrt((np.tanh(c * Xs*sigmas)**2).mean(0)), label = f'Offset = {c:.1f}')

plt.plot(sigmas, sigmas, 'k--', alpha = 0.5, label = 'Identity')
plt.xlabel(r'$\sigma_i$')
plt.ylabel(r'$\sigma_{i+1}$')
#plt.legend(bbox_to_anchor = (1., 1.))
plt.legend()
plt.tight_layout()
plt.savefig(path + 'tanh_offset.pdf', dpi = 200)
# plt.xscale('log')
# plt.yscale('log')
plt.show()
plt.close()
```

```python
X = np.array(onp.random.randn(4096, *sigmas.shape))
plt.figure(figsize = (4,3))

mapping = std_mapping(np.tanh)
map_grad = grad(mapping)

for offset in [1., 2., 3.]:

    map_dev = vmap(map_grad, in_axes = (0, 1, None))(sigmas, X, offset)
    plt.plot(sigmas, map_dev, label = f'Offset = {offset:.1f}')
plt.axhline(1, linestyle = '--', color = 'k', alpha = 0.5)
plt.legend(bbox_to_anchor = (1., 1.))
plt.xlabel(r'$\sigma_i$')
plt.tight_layout()
plt.savefig(path + 'rhs_derivative_offset.pdf', dpi = 200)
plt.show()
plt.close()
```

```python
def rhs_fn(x, offset):
    return np.interp(x, sigmas, np.sqrt((np.tanh(offset * Xs*sigmas)**2).mean(0)))

def find_root(fun, max_it = 200, minval = 0.01, maxval = 4, eps = 1e-2):
    
    
    for i in range(max_it):
        midval = (maxval + minval) / 2
        
        max_f = fun(maxval)
        min_f = fun(minval)
        mid_f = fun(midval)
        
        if mid_f * max_f < 0:
            minval = midval
        else:
            maxval = midval

        
        if np.abs((maxval - minval)) < eps:
#             print('Maximum iteration not reached')
            break
    return midval

def rhs_root(offset, **kwag):
    fun = lambda x: rhs_fn(x, offset) - x
    return find_root(fun, **kwag)
```

```python
offsets = np.linspace(1, 10, 50)

rhs_roots = []
rhs_root_grads = []
zero_grads = []
for offset in offsets:
    root = rhs_root(offset)
    rhs_roots.append(root)
    rhs_root_grads.append(map_grad(root, X[:,0], offset))
    zero_grads.append(map_grad(0.01, X[:,0], offset))
```

```python
plt.figure(figsize = (6,3))
plt.subplot(1,2,1)
plt.plot(offsets, rhs_roots)
plt.xlabel(r'Offset')
plt.ylabel('Positive fixed point')
plt.tight_layout()
plt.subplot(1,2,2)
plt.plot(offsets, rhs_root_grads, label = 'Positive')
plt.plot(offsets, zero_grads, label = 'Zero')
plt.yscale('log')
plt.xlabel(r'Offset')
plt.ylabel('Derivative at the fixed point')
plt.legend()
plt.tight_layout()

plt.savefig(path + f'rhs_roots_and_grad.pdf', dpi = 200)

plt.show()
plt.close()
```
