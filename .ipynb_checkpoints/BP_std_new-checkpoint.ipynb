```python
from jax import numpy as np
from jax import grad, vmap, random
import numpy as onp
from matplotlib import pyplot as plt
```

```python
path = 'dynamical_system/bp_new_'
```

```python
plt.rcParams.update(plt.rcParamsDefault)
plt.rc('font', family='serif')
plt.rcParams.update({'font.size': 16,
                        'lines.linewidth': 2,
                        'axes.labelsize': 10,
                        'axes.titlesize': 16,
                        'xtick.labelsize': 10,
                        'ytick.labelsize': 10,
                        'legend.fontsize': 10,
                        'axes.linewidth': 2})
```

```python
sigma = np.linspace(0,4,100)
X = np.array(onp.random.randn(4096, 100)).T
```

```python
rhs = lambda sigma, X, offset: np.sqrt(((1 - np.tanh(offset * X * sigma) ** 2)**2).mean())
```

```python
plt.figure(figsize = (6,3))
plt.subplot(1,2,1)
for offset, c in zip([1., 4., 8.], ['r', 'g', 'b']):
    plt.plot(sigma, vmap(rhs, in_axes = [0, 0, None])(sigma, X, offset), color = c, label = f'offset = {int(offset)}', alpha = 0.8)
    #plt.plot(sigma, vmap(grad(rhs), in_axes = [0, 0, None])(sigma, X, offset), linestyle = '--', color = c, alpha = 0.8)
plt.plot(sigma, sigma, 'k-.', linewidth = 1)
# plt.axhline(-1, color = 'k', linestyle = '--', linewidth = 1)
plt.xlabel(r'$s^T_X$')
plt.ylabel(r'$s^T_Y$')
plt.xlim([0, 2])
plt.ylim([0,2])
plt.legend()
plt.tight_layout()

plt.subplot(1,2,2)

for offset, c in zip([1., 4., 8.], ['r', 'g', 'b']):
    #plt.plot(sigma, vmap(rhs, in_axes = [0, 0, None])(sigma, X, offset), color = c, label = f'offset = {int(offset)}', alpha = 0.8)
    plt.plot(sigma, vmap(grad(rhs), in_axes = [0, 0, None])(sigma, X, offset), linestyle = '--', color = c, alpha = 0.8, label = f'offset = {int(offset)}')
# plt.plot(sigma, sigma, 'k-.', linewidth = 1)
plt.axhline(-1, color = 'k', linestyle = '--', linewidth = 1)
plt.xlabel(r'$s^T_X$')
plt.ylabel(r'$ds^T_Y/ds^T_X$')
plt.xlim([0, 2])
# plt.ylim([-3,2])
plt.legend()
plt.tight_layout()

plt.savefig(path + f'rhs_multi_offset.pdf', dpi = 200)
plt.show()
plt.close()
```

```python
def rhs_fn(x, offset, S = sigma, X = X):
    return np.interp(x, S, vmap(rhs, in_axes = [0, 0, None])(S, X, offset))

def find_root(fun, max_it = 200, minval = 0.01, maxval = 4, eps = 1e-2):
    
    
    for i in range(max_it):
        midval = (maxval + minval) / 2
        
        max_f = fun(maxval)
        min_f = fun(minval)
        mid_f = fun(midval)
        
        if mid_f * max_f < 0:
            minval = midval
        else:
            maxval = midval

        
        if np.abs((maxval - minval)) < eps:
#             print('Maximum iteration not reached')
            break
    return midval

def rhs_root(offset, **kwag):
    fun = lambda x: rhs_fn(x, offset) - x
    return find_root(fun, **kwag)

def rhs_grad_fn(x, offset, S = sigma, X = X):
    return np.interp(x, S, vmap(grad(rhs), in_axes = [0, 0, None])(S, X, offset))
    
```

```python
offsets = np.linspace(0.01, 20, 100)

rhs_roots = []
rhs_root_grads = []
for offset in offsets:
    root = rhs_root(offset)
    rhs_roots.append(root)
    rhs_root_grads.append(rhs_grad_fn(root, offset))
    
```

```python
plt.figure(figsize = (6,3))
plt.subplot(1,2,1)
plt.plot(offsets, rhs_roots)
plt.xlabel(r'Offset')
plt.ylabel('Fixed point')
plt.tight_layout()
plt.subplot(1,2,2)
plt.plot(offsets, rhs_root_grads)
plt.xlabel(r'Offset')
plt.ylabel('Derivative at the fixed point')
plt.tight_layout()

plt.savefig(path + f'rhs_roots_and_grad.pdf', dpi = 200)

plt.show()
plt.close()
```
